```python
import requests
```

```python
#title = "computer" 
#author = "Jobs" 
url = "http://www.patentsview.org/api/patents/query?q" 
data = { 
    "q":{"_gte":{"patent_date":"2019-01-01"}}, 
    #"o":{"matched_subentities_only":"true"} 
} 
resp = requests.post(url, json=data) 
```

```python
with open("patents.txt", "w") as f:
    f.write(resp.text)
```
